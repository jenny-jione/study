## 2.1 재귀함수

```python
import time
def countdown(n):
    while True:
        if n == 0:
            print("happy new year!!")
            break
        print(n)
        time.sleep(0.5)
        n -= 1
```

```python
countdown(10)
```

```python
def countdown2(n):
    if n == 0:
        print("happy new new year!")
    else:
        print(n)
        time.sleep(0)
        countdown2(n-1)
```

```python
countdown2(10)
```

```python
countdown2(10000000)
```

RecursionError: maximum recursion depth exceeded while calling a Python object

countdown2 는 에러가 나지만 countdown은(while문으로 만든) 에러 없이 계속된다.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(10)
```

```python
def loop_factorial(n):
    num = 1
    while n >= 1:
        num = num * n
        n = n - 1
    return num
```

```python
loop_factorial(10)
```

```python
from collections.abc import MutableMapping 

def to_complete_flat_dict(dictionary, parent_key=False, separator='.'):
    items = []
    for key, value in dictionary.items():
        new_key = str(parent_key) + separator + key if parent_key else key
        if isinstance(value, MutableMapping): #dictionary인지 탐색
            items.extend(to_complete_flat_dict(value, new_key, separator).items())
        elif isinstance(value, list):
            for k, v in enumerate(value):
                items.extend(to_complete_flat_dict({str(k):v}, new_key).items())
        else:
            items.append((new_key, value))
    return dict(items)
```

```python
to_complete_flat_dict(your_data)
```
