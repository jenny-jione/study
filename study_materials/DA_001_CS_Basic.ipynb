# 1. 컴퓨터의 기본적 구조
* 모든 프로그래밍 언어를 제대로 알기 전에 컴퓨터의 대략적인 구조와 그 원리를 알아야 원활한 코딩이 가능합니다.
* 프로그래밍 언어는 기본적으로 컴퓨터를 조작하기 위한 언어이기 때문입니다.

## 1.1 컴퓨터의 구조

* 크게 3가지 부분으로 이루워져 있습니다.
    1. 중앙처리장치 (CPU)
    2. 기억장치 (RAM)
    3. 입출력장치 (I/O)
* 위 3가지 장치중 하나라도 없으면 컴퓨터는 명령을 처리할 수 없으며, 컴퓨터라 할 수 없습니다.

**CPU**
* CPU는 명령어를 연산하는 기능을 수행합니다.
* 우리가 쓴 코드는 CPU를 통해서 처리되어 원하는 결과를 얻게끔 해 줍니다. <br>

**기억장치**
* 기억장치 (RAM)은 CPU가 처리할 명령어, 변수, 상수, 코드, 등을 보관하고, CPU가 이를 불러올 수 있는 기능을 합니다.
* RAM에는 크게 코드영역, 데이터 영역, 스택영역, 힙 영역 네가지가 있습니다.
* 이 중 힙 영역이 우리 프로그래머들이 관리하고 사용하는 영역이라고 생각하면 됩니다.
* 이 힙 영역에 우리가 만든 자료형, 데이터들이 저장되고, 이 힙 주소를 스택 영역에 저장한다 생각하시면 됩니다.
* 프로그램이 종료될 시 모든 메모리 값들이 삭제됩니다. <br>

**입출력장치**
* 입출력장치는 명령어를 입력, 출력의 결과를 확인하는 기능을 합니다.
* 키보드와 모니터가 대표적인 입력, 출력 장치입니다.

## 1.2 프로그래밍 언어
* 컴퓨터는 전기 신호로 움직이는 일종의 엔진입니다.
* 자동차가 휘발류로 엔진을 돌려 운전자가 조작을 통해 가속, 감속, 라디오틀기 등의 기능을 수행하는것 처럼,
* 컴퓨터또한 프로그래밍 언어를 통해 컴퓨터의 전기신호를 제어하여 우리가 원하는 결과를 받는것이 코딩입니다.<br>

**저수준언어 vs 고수준언어**
<img src='./DA_001/1.png' width="50%" height="50%">
* 컴퓨터는 기본적으로 0, 1의 값만 처리할 수 있습니다.
* 하지만 인간은 0과 1만 사용하여 프로그래밍 하기엔 무리가 있습니다.
* 그렇기때문에 인간과 친화적인 언어로 프로그래밍 하고, 그 언어를 0 과 1로 번역하여 컴퓨터가 알아들을 수 있도록 하는것이 현재 프로그래밍입니다.
* 파이썬, 자바와 같은 인간친화적인 언어가 고수준언어이며, 어셈블리어, 0, 1 이진수 같은 컴퓨터 친화적인 언어가 저수준 언어입니다.
* 고수준의 언어를 컴퓨터가 알아들을 수 있게끔 번역해주는것이 컴파일러 입니다.

## 1.3 OS (운영체제)
* 초기의 컴퓨터는 검은색 바탕화면만 있고, 키보드로 입력한 연산 결과를 출력해주는 기본적인 동작만 하였습니다.
* 이것을 터미널이라 불렀으며, 사람과 컴퓨터를 연결해주는 통로 역할이라는 의미로 위와같은 이름이 붙여졌습니다.
* 이것이 최초로 컴퓨터를 제어하는 운영체제 이며, 계산기를 생각하면 쉽겠습니다.
* 이것을 CLI (Command Line Interface) 라 부르며, 커맨드를 입력하여 컴퓨터의 연산결과를 받는다는 의미로 이런 이름이 붙어졌습니다.
* 하지만 CLI 는 사용자 친화적이지 않았습니다. 그래서 등장한것이 GUI (Graphic User Interface) 입니다.
* 현재 우리가 사용하는 윈도우, 맥, 안드로이드, 리눅스 등 모든 운영체제는 GUI 이며, 키보드와 마우스를 사용한 시각적인 상호반응을 사용하여 커맨드가 자동으로 입력되며, 그 연산결과 또한 시각적으로 받고 있습니다. <br>
* 즉 운영체제는 사용자와 컴퓨터의 상호작용을 도와주는 프로그램이라 생각하시면 됩니다.

**운영체제 안에서의 터미널**
* 모든 운영체제는 자체 터미널을 가지고 있습니다.
* GUI 운영체제 안에 CLI 인 터미널을 사용한다는 의미는 컴퓨터를 시각적으로 조작하는것이 아니라, 명령어로 조작한다는 의미입니다.
* GUI 에서 마우스클릭으로 컴퓨터에게 명령을 내리는것을 터미널을 사용하여 명령하겠다고 생각하시면 되겠습니다. <br>

### 1.3.1 Shell (쉘)
* 쉘 (혹은 쉘스크립트)은 커널(하드웨어를 조작하는 운영체제의 프로그램)과 사용자 사이에서 사용자의 명령을 해석하고 컴퓨터를 조작하는 시스템 프로그램입니다.
* 쉘 스크립트 또한 파이썬과 같은 컴퓨터 프로그래밍 언어이고, 터미널에서 사용되는 언어라고 생각하면 쉽겠습니다.
* 각 운영체제 마다 CLI에서 컴퓨터를 조작하기 위해 기본적으로 내장된 프로그램이 쉘입니다.
* 터미널에서 파이썬을 사용하고 싶다는 의미는 컴퓨터에 파이썬을 설치하여, 쉘처럼 명령을 내리겠다 라는 말입니다.
