# 1. 파이썬 루프 

* 루프 (반복문)은 프로그래밍의 기본이고 알파이자 오메가 입니다.
* for, while을 사용할수 있다는 가정에서 시작하겠습니다.

##  1.1 for 문

* 가장 자주 접하는 반복문입니다.
* iterable 한 객체를 반복시켜 줍니다.
* iterable 객체: list, tuple, dictionary, generator, set, str, bytes, range 등등..

````python
# 기본 사용법
for i in range(0, 10):
    print(i)
````

````python
# 응용 사용법
for i in range(0, 10):
    if i % 3 == 0:
        print(i)
````

````python
# 응용 사용법 2
for i in range(0, 2990):
    if i % 3 == 0:
        print(i)
    elif i == 10:
        break
    else:
        print('no!!!')
````

````python
# 응용 사용법 3
list_x = ['1','2','3','4','5']
list_y = ['a','b','c','d','e']

for i, (x, y) in enumerate(zip(list_x, list_y)):
    print(i ,x, y)
    # print(list_x[i], list_y[i+1])
````

````python
# 실전 예제
your_list = range(0, 1000, 3)
index = 0
for i in range(index, len(your_list)):
    print(your_list[i])
````

````python
index = int(input())
print('*'*30)
for i in range(index, len(your_list)):
    print(your_list[i])
````

## 1.2 while 문

* while 문은 생각보다 실제 컴퓨터에서 매우 자주 사용되고 있습니다.
* 우리가 보고 있는 인터넷 브라우저가 항상 켜져 있는게 신기하지 않나요?
* 그 이유는 while 문으로 브라우저 창이 계속 화면에 표시되기 때문입니다.
* 그리고 사용자가 x 버튼을 눌러 while 문을 빠져 나가도록 코딩이 되어 있습니다.
* 즉 while문은 특정 조건이 성립될때까지 계속 반복을 하는 코드에 알맞습니다.
* while문을 사용할때 항상 루프 문을 빠져나올 수 있도록 코딩하도록 합시다.

````python
# while문 기본
i = 0
while i != 21:
    print(i)
    i += 1
````

````python
# while문 기본 2
i = 0
while True:
    print(i)
    if i == 20:
        break
    i += 1
````

````python
# while 문 실전 응용
while True:
    print(f'Select your choice:: 1. print something, 2. print nothing, 3. exit')
    choice = int(input('Your choice? -> ' ))
    if choice == 1:
        print('something\n')
    elif choice == 2:
        print('nothing\n')
    elif choice == 3:
        break
    else:
        print('Wrong command\n')
````

````python
# while 문 실전 응용
import random

count = 0
number = random.randint(1, 10002)
print(f'Target number: {number}')

while True:
    number2 = random.randint(1, 10002)
    if number == number2:
        break
    count += 1
print(f'Numbers of tries: {count}')
````

# 2. 파이썬 예외처리 (Exception)

* 모든 프로그래밍 언어는 예상치 못한 에러를 처리하기 위한 예외 (Exception) 처리 기능이 있습니다.
* Java 는 ```try, catch``` 문이고, 우리 Python 은 ```try, except``` 입니다. 
* 소규모 작은 프로젝트는 if, elif, else 의 분기문을 사용하여 예상되는 기능을 처리할 수 있지만,
* 프로젝트나 코드의 규모가 커지면 커질수록 개발자 (혹은 분석가!) 가 예상하지 못한 에러가 발생할 수 있습니다.,
* 그러한 에러를 잡아내어 코드의 안전성을 보장하는것이 예외처리입니다.

````python
# 예외처리 기본
try:
    their_list = range(0, 10)
    index = int(input('Entered number is: '))
    print(their_list[index])

except IndexError:
    index =0
    print(their_list[index])

except Exception as e:
    print(f'Error! Original Error: {e}')
````
